```python
# NAME: roster.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
#add items to list
#remove items from list
#Edit item on the list
# PROGRAM LOGIC
#While True loop
#see the list
#call see the list fuction
#add the player to the list
#call the add item fuction
#ask user which player to quit
#edit item function
#Call add_players to fill the list.
#Ask the user for one specific name to search for.
#Call is_on_roster and use the returned value to print a message
# Function 1: add_players
def add_players(player_list):
    while True:
        name = input("Enter player name (or 'quit' to stop): ")
        if name.lower() == "quit":
            break
        player_list.append(name)


# Function 2: is_on_roster
def is_on_roster(player_list, search_name):
    return search_name in player_list


# Program Logic
roster = []

add_players(roster)

search_name = input("\nWho would you like to search for? ")

if is_on_roster(roster, search_name):
    print("Result: Player found!")
else:
    print("Result: Player not found.")
```
