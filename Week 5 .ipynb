```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
def calculate_ticket(age, hour):
    if age <= 12 or age >= 65:
        price = 8
    else:
        price = 12

    # For testing age logic with a standard hour, we'll print the price here
    # The hour-based logic will be added later.
    print("Calculated price for age {age} (at {hour}:00) is: ${price}")
    return price

# PROGRAM LOGIC
# Test with different ages and a standard hour (19:00)
print("Testing age-based pricing:")
calculate_ticket(age=10, hour=19)  # Should be $8
calculate_ticket(age=30, hour=19)  # Should be $12
calculate_ticket(age=70, hour=19)  # Should be $8
```
