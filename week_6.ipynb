```python
# IMPORT STATEMENTS
# 1. Import math

# FUNCTION DEFINITIONS
# 1. Define area(radius): returns pi * r squared
# 2. Define circumference(radius): returns 2 * pi * r
# 3. Define volume(radius, height):
#    - Call area(radius) and multiply by height.
#    - Return the result.

# PROGRAM LOGIC
# 1. Input for radius (float)
# 2. Call and print area and circumference.
# 3. Input for height (float)
# 4. Call and print volume.
import math

def area(radius):
    return math.pi * (radius ** 2)

def circumference(radius):
    return 2 * math.pi * radius

# Get user input for radius
radius_input = float(input("Enter the radius of the cylinder: "))

# Calculate and print area and circumference
calculated_area = area(radius_input)
calculated_circumference = circumference(radius_input)

print(f"Area: {calculated_area}")
print(f"Circumference: {calculated_circumference}")

```

## Define and Test Volume

### Subtask:
Define the `volume(radius, height)` function, which should call the `area(radius)` function internally. Collect user input for the height as a float. Call the `volume` function with the previously entered radius and the new height, and print the result to verify accuracy.


**Reasoning**:
The subtask requires defining the `volume` function, taking user input for the height, and then calling the `volume` function with the previously obtained radius and the new height to print the result. This can be achieved with a single Python code block by extending the existing code.



```python
import math

def area(radius):
    return math.pi * (radius ** 2)

def circumference(radius):
    return 2 * math.pi * radius

def volume(radius, height):
    return area(radius) * height

# Get user input for radius
radius_input = float(input("Enter the radius of the cylinder: "))

# Calculate and print area and circumference
calculated_area = area(radius_input)
calculated_circumference = circumference(radius_input)

print(f"Area: {calculated_area}")
print(f"Circumference: {calculated_circumference}")

# Get user input for height
height_input = float(input("Enter the height of the cylinder: "))

# Calculate and print volume
calculated_volume = volume(radius_input, height_input)
print(f"Volume: {calculated_volume}")
```
