```python
# NAME: rentalcar.py
# IMPORT STATEMENTS
# FUNCTION DEFINITIONS
# Rental Car Pricing System
# Function to return the full name of the classification
def get_classification():
    code = input("Enter car classification (B = Budget, S = Standard, L = Luxury): ").upper()

    while code not in ['B', 'S', 'L']:
        print("Invalid classification. Please enter B, S, or L.")
        code = input("Enter car classification (B = Budget, S = Standard, L = Luxury): ").upper()

    return code

def get_class_name(code):
    if code == 'B':
        return "Budget"
    elif code == 'S':
        return "Standard"
    elif code == 'L':
        return "Luxury"
# Function to get the daily rate
def get_daily_rate(code):
    if code == 'B':
        return 40
    elif code == 'S':
        return 60
    elif code == 'L':
        return 80
# Function to calculate miles charged
def calculate_miles_charged(miles):
    if miles > 100:
        return miles - 100
    else:
        return 0
# MAIN BILLING FUNCTION
def calculate_total_bill(days, code, miles):
    rate = get_daily_rate(code)
    base_cost = days * rate

    miles_charged = calculate_miles_charged(miles)
    mileage_cost = miles_charged * 0.35

    total = base_cost + mileage_cost

    return total, miles_charged

# PROGRAM Logic
days = int(input("Enter number of rental days: "))
classification = get_classification()
miles_driven = int(input("Enter total miles driven: "))

# Calculate bill
total_bill, charged_miles = calculate_total_bill(days, classification, miles_driven)
# Display results
print("\n----- Rental Summary -----")
print("Car Classification:", get_class_name(classification))
print("Total Miles Driven:", miles_driven)
print("Miles Charged:", charged_miles)
print("Total Bill: $", format(total_bill, ".2f"))
```
