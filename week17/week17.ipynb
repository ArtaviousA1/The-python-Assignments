```python
tasks = []
HISTORY_FILE = "history.txt"

def add_task():
    name = input("Enter task name: ")
    try:
        priority = int(input("Enter priority (1=High, 2=Medium, 3=Low): "))
        if priority not in [1, 2, 3]:
            print("Invalid priority.")
            return
    except ValueError:
        print("Priority must be a number.")
        return

    tasks.append({"name": name, "priority": priority})
    print("Task added.\n")

def view_tasks():
    if not tasks:
        print("No tasks available.\n")
        return

    print("\nYour Tasks:")
    for i, task in enumerate(tasks):
        print(f"{i + 1}. {task['name']} (Priority {task['priority']})")
    print()

def complete_task():
    view_tasks()
    if not tasks:
        return

    try:
        index = int(input("Enter task number to complete: ")) - 1
        task = tasks.pop(index)

        with open(HISTORY_FILE, "a") as f:
            f.write(task["name"] + "\n")

        print(f"Task '{task['name']}' completed and saved.\n")
    except (ValueError, IndexError):
        print("Invalid selection.\n")

def what_is_next():
    for task in tasks:
        if task["priority"] == 1:
            print(f"\nNext High Priority Task: {task['name']}\n")
            return

    print("\nNo high-priority tasks 🎉 You're doing great, keep going!\n")

def main():
    while True:
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Complete Task")
        print("4. What is next?")
        print("5. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            complete_task()
        elif choice == "4":
            what_is_next()
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice.\n")

if __name__ == "__main__":
    main()
```
