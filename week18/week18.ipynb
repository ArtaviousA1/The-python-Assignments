```python
# PART 1: STRUCTURE & PSEUDOCODE

# START PROGRAM

# Create empty list called expenses

# WHILE user has not chosen to exit:
#     Display menu:
#         1. Add New Expense
#         2. Get Totals
#         3. Exit

#     Get user selection

#     IF selection == 1:
#         Ask for category
#         Ask for vendor
#         Ask for amount
#         Ask for date
#
#         Create a dictionary:
#             {"category": category,
#              "vendor": vendor,
#              "amount": amount,
#              "date": date}
#
#         Add dictionary to expenses list

#     ELSE IF selection == 2:
#         Show totals menu:
#             A: All expenses
#             C: By category
#             V: By vendor
#
#         Get choice
#
#         IF A:
#             Sum all expense amounts
#
#         IF C:
#             Ask for category
#             Loop through expenses and sum matching category
#
#         IF V:
#             Ask for vendor
#             Loop through expenses and sum matching vendor
#
#         Display total formatted as money

#     ELSE IF selection == 3:
#         End program

# END PROGRAM
expenses = []

while True:
    print("\n--- Expense Tracker ---")
    print("1. Add New Expense")
    print("2. Get Totals")
    print("3. Exit")

    choice = input("Selection: ")

    if choice == "1":
        category = input("Enter Category: ")
        vendor = input("Enter Vendor: ")

        while True:
            try:
                amount = float(input("Enter Amount: "))
                break
            except ValueError:
                print("Please enter a valid number.")

        date = input("Enter Date: ")

        expense = {
            "category": category,
            "vendor": vendor,
            "amount": amount,
            "date": date
        }

        expenses.append(expense)
        print("Expense added!")

    elif choice == "2":
        print("\n--- Totaling Options ---")
        print("A: All Expenses")
        print("C: Filter by Category")
        print("V: Filter by Vendor")

        option = input("Choice: ").upper()

        total = 0

        if option == "A":
            for exp in expenses:
                total += exp["amount"]

            print(f"Total of all expenses: ${total:.2f}")

        elif option == "C":
            category = input("Enter Category to sum: ")

            for exp in expenses:
                if exp["category"].lower() == category.lower():
                    total += exp["amount"]

            print(f"The total for {category} is: ${total:.2f}")

        elif option == "V":
            vendor = input("Enter Vendor to sum: ")

            for exp in expenses:
                if exp["vendor"].lower() == vendor.lower():
                    total += exp["amount"]

            print(f"The total for {vendor} is: ${total:.2f}")

        else:
            print("Invalid option.")

    elif choice == "3":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
