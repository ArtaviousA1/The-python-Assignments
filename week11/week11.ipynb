```python
# NAME: gradebook.py
# IMPORT STATEMENTS
# while loop to collect student names
# stop when user types "done"
# store name and grade in dictionary
# calculate average using sum of values
# FUNCTION DEFINITIONS
# PROGRAM LOGIC
# NAME: gradebook.py

# FUNCTION DEFINITIONS

def add_students(gradebook):
    while True:
        name = input("Enter student name (or 'done'): ")

        if name.lower() == "done":
            break

        grade = float(input(f"Enter {name}'s grade: "))
        gradebook[name] = grade


def calculate_average(gradebook):
    if len(gradebook) == 0:
        return 0

    total = sum(gradebook.values())
    return total / len(gradebook)


# PROGRAM LOGIC

gradebook = {}

add_students(gradebook)

print("\n--- Class Report ---")

for student, grade in gradebook.items():
    print(f"{student}: {grade}")

average = calculate_average(gradebook)

print(f"Class Average: {average:.2f}")
```
