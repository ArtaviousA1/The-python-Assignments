```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
def calculate_ticket(age, hour):
    if age <= 12 or age >= 65:
        price = 8
    else:
        price = 12

    # Apply hour-based discount
    if hour < 17:
        price -= 2

    print(f"Your ticket price is ${price}.00")
    return price

# PROGRAM LOGIC
# Get user input for age and hour
age_input = int(input("Enter age: "))
hour_input = int(input("Enter movie hour (0-23): "))

# Call calculate_ticket with user input
calculate_ticket(age_input, hour_input)
```
