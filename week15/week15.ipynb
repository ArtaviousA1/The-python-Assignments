```python
import json
import os

TASKS_FILE = "tasks.json"

def load_tasks():
    if not os.path.exists(TASKS_FILE):
        return []
    with open(TASKS_FILE, "r") as f:
        return json.load(f)

def save_tasks(tasks):
    with open(TASKS_FILE, "w") as f:
        json.dump(tasks, f, indent=4)

def add_task():
    name = input("Enter task name: ")
    priority = input("Enter priority (1=High, 2=Medium, 3=Low): ")

    if priority not in ["1", "2", "3"]:
        print("Invalid priority.")
        return

    tasks = load_tasks()
    tasks.append({
        "name": name,
        "priority": int(priority)
    })
    save_tasks(tasks)
    print("Task added.")

def view_tasks():
    tasks = load_tasks()
    if not tasks:
        print("No tasks found.")
        return

    for i, task in enumerate(tasks):
        print(f"{i+1}. {task['name']} (Priority {task['priority']})")

def main():
    while True:
        print("\n1. Add Task")
        print("2. View Tasks")
        print("3. Exit")

        choice = input("Choose: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
