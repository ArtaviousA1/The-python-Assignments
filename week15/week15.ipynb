```python
import json
import os

TASKS_FILE = "tasks.json"
HISTORY_FILE = "history.json"

def load_data(file):
    if not os.path.exists(file):
        return []
    with open(file, "r") as f:
        return json.load(f)

def save_data(file, data):
    with open(file, "w") as f:
        json.dump(data, f, indent=4)

def add_task():
    name = input("Enter task name: ")
    priority = input("Enter priority (1=High, 2=Medium, 3=Low): ")

    if priority not in ["1", "2", "3"]:
        print("Invalid priority.")
        return

    tasks = load_data(TASKS_FILE)
    tasks.append({
        "name": name,
        "priority": int(priority)
    })
    save_data(TASKS_FILE, tasks)
    print("Task added.")

def view_tasks():
    tasks = load_data(TASKS_FILE)
    if not tasks:
        print("No tasks found.")
        return

    for i, task in enumerate(tasks):
        print(f"{i+1}. {task['name']} (Priority {task['priority']})")

def complete_task():
    tasks = load_data(TASKS_FILE)

    if not tasks:
        print("No tasks to complete.")
        return

    view_tasks()
    choice = int(input("Enter task number to complete: ")) - 1

    if choice < 0 or choice >= len(tasks):
        print("Invalid choice.")
        return

    completed = tasks.pop(choice)
    save_data(TASKS_FILE, tasks)

    history = load_data(HISTORY_FILE)
    history.append(completed)
    save_data(HISTORY_FILE, history)

    print("Task completed and saved to history.")

def main():
    while True:
        print("\n1. Add Task")
        print("2. View Tasks")
        print("3. Complete Task")
        print("4. Exit")

        choice = input("Choose: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            complete_task()
        elif choice == "4":
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
