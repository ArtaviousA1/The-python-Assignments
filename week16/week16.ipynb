```python
import json
import os

DATA_FILE = "stats.json"


class Game:
    def __init__(self, opponent, points, rebounds, assists):
        self.opponent = opponent
        self.points = points
        self.rebounds = rebounds
        self.assists = assists

    def to_dict(self):
        return self.__dict__


class Player:
    def __init__(self):
        self.games = []

    # -------- INPUT HELPERS --------
    def get_int(self, prompt):
        while True:
            value = input(prompt).strip()
            if not value:
                print("❌ Input cannot be empty.")
                continue
            try:
                num = int(value)
                if num < 0:
                    print("❌ Cannot be negative.")
                    continue
                return num
            except ValueError:
                print("❌ Please enter a valid number.")

    def get_text(self, prompt):
        while True:
            text = input(prompt).strip()
            if text:
                return text
            print("❌ Text cannot be empty.")

    # -------- CORE FEATURES --------
    def add_game(self):
        opponent = self.get_text("Opponent team: ")
        points = self.get_int("Points scored: ")
        rebounds = self.get_int("Rebounds: ")
        assists = self.get_int("Assists: ")

        self.games.append(Game(opponent, points, rebounds, assists))
        print("✅ Game added!\n")

    def show_games(self):
        if not self.games:
            print("No games recorded.\n")
            return

        for i, g in enumerate(self.games, 1):
            print(f"{i}. {g.opponent}: {g.points} pts, {g.rebounds} reb, {g.assists} ast")

    def show_averages(self):
        if not self.games:
            print("No data to calculate averages.\n")
            return

        total_games = len(self.games)
        total_points = sum(g.points for g in self.games)
        total_rebounds = sum(g.rebounds for g in self.games)
        total_assists = sum(g.assists for g in self.games)

        print("\n📊 Averages Per Game:")
        print(f"Points: {total_points / total_games:.2f}")
        print(f"Rebounds: {total_rebounds / total_games:.2f}")
        print(f"Assists: {total_assists / total_games:.2f}\n")

    def delete_game(self):
        self.show_games()
        if not self.games:
            return

        index = self.get_int("Enter game number to delete: ") - 1

        if 0 <= index < len(self.games):
            confirm = input("Are you sure? (y/n): ").lower()
            if confirm == "y":
                removed = self.games.pop(index)
                print(f"🗑 Removed game vs {removed.opponent}")
        else:
            print("❌ Invalid selection.")

    # -------- FILE HANDLING --------
    def save_data(self):
        try:
            with open(DATA_FILE, "w") as f:
                json.dump([g.to_dict() for g in self.games], f, indent=4)
            print(f"💾 Saved to {DATA_FILE}")
        except Exception as e:
            print("❌ Error saving file:", e)

    def load_data(self):
        if not os.path.exists(DATA_FILE):
            return

        try:
            with open(DATA_FILE, "r") as f:
                data = json.load(f)
                self.games = [Game(**g) for g in data]
        except Exception:
            print("⚠️ Warning: Data file corrupted. Starting fresh.")
            self.games = []


# -------- MAIN APP --------
def main():
    player = Player()
    player.load_data()

    while True:
        print("\n--- Basketball Tracker ---")
        print("1. Add Game")
        print("2. Show Games")
        print("3. Show Averages")
        print("4. Delete Game")
        print("5. Save")
        print("6. Exit")

        choice = input("Choose: ").strip()

        if choice == "1":
            player.add_game()
        elif choice == "2":
            player.show_games()
        elif choice == "3":
            player.show_averages()
        elif choice == "4":
            player.delete_game()
        elif choice == "5":
            player.save_data()
        elif choice == "6":
            player.save_data()
            print("👋 Goodbye!")
            break
        else:
            print("❌ Invalid option.")


if __name__ == "__main__":
    main()
```
