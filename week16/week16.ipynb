```python
import json
import os

DATA_FILE = "stats.json"


class Game:
    def __init__(self, opponent, points, rebounds, assists):
        self.opponent = opponent
        self.points = points
        self.rebounds = rebounds
        self.assists = assists

    def to_dict(self):
        return self.__dict__


class Player:
    def __init__(self):
        self.games = []

    def add_game(self):
        opponent = input("Opponent team: ")
        points = int(input("Points scored: "))
        rebounds = int(input("Rebounds: "))
        assists = int(input("Assists: "))

        game = Game(opponent, points, rebounds, assists)
        self.games.append(game)

    def show_games(self):
        for g in self.games:
            print(f"{g.opponent}: {g.points} pts, {g.rebounds} reb, {g.assists} ast")

    def save_data(self):
        with open(DATA_FILE, "w") as f:
            json.dump([g.to_dict() for g in self.games], f)

    def load_data(self):
        if os.path.exists(DATA_FILE):
            with open(DATA_FILE, "r") as f:
                data = json.load(f)
                for g in data:
                    self.games.append(Game(**g))


def main():
    player = Player()
    player.load_data()

    while True:
        print("\n1. Add Game")
        print("2. Show Games")
        print("3. Save & Exit")

        choice = input("Choose: ")

        if choice == "1":
            player.add_game()
        elif choice == "2":
            player.show_games()
        elif choice == "3":
            player.save_data()
            print("Saved!")
            break
        else:
            print("Invalid option")


if __name__ == "__main__":
    main()
```
