```python
import json
import os

DATA_FILE = "stats.json"


class Game:
    def __init__(self, opponent, points, rebounds, assists):
        self.opponent = opponent
        self.points = points
        self.rebounds = rebounds
        self.assists = assists

    def to_dict(self):
        return self.__dict__


class Player:
    def __init__(self):
        self.games = []

    def add_game(self):
        try:
            opponent = input("Opponent team: ")
            points = int(input("Points scored: "))
            rebounds = int(input("Rebounds: "))
            assists = int(input("Assists: "))

            game = Game(opponent, points, rebounds, assists)
            self.games.append(game)
            print("Game added!\n")

        except ValueError:
            print("❌ Please enter numbers for stats.\n")

    def show_games(self):
        if not self.games:
            print("No games recorded.\n")
            return

        for i, g in enumerate(self.games, 1):
            print(f"{i}. {g.opponent}: {g.points} pts, {g.rebounds} reb, {g.assists} ast")

    def show_averages(self):
        if not self.games:
            print("No data to calculate averages.\n")
            return

        total_games = len(self.games)
        total_points = sum(g.points for g in self.games)
        total_rebounds = sum(g.rebounds for g in self.games)
        total_assists = sum(g.assists for g in self.games)

        print("\n📊 Averages Per Game:")
        print(f"Points: {total_points / total_games:.2f}")
        print(f"Rebounds: {total_rebounds / total_games:.2f}")
        print(f"Assists: {total_assists / total_games:.2f}\n")

    def save_data(self):
        with open(DATA_FILE, "w") as f:
            json.dump([g.to_dict() for g in self.games], f, indent=4)
        print(f"💾 Data saved to {DATA_FILE}")

    def load_data(self):
        if os.path.exists(DATA_FILE):
            with open(DATA_FILE, "r") as f:
                data = json.load(f)
                self.games = [Game(**g) for g in data]


def main():
    player = Player()
    player.load_data()

    while True:
        print("\n--- Basketball Tracker ---")
        print("1. Add Game")
        print("2. Show Games")
        print("3. Show Averages")
        print("4. Save")
        print("5. Exit")

        choice = input("Choose: ")

        if choice == "1":
            player.add_game()
        elif choice == "2":
            player.show_games()
        elif choice == "3":
            player.show_averages()
        elif choice == "4":
            player.save_data()
        elif choice == "5":
            player.save_data()
            print("Goodbye!")
            break
        else:
            print("Invalid option\n")


if __name__ == "__main__":
    main()
```
