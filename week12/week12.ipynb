```python
# NAME: rectangle.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# Function: calculate_rectangle(length, width)
# - Take in length and width as parameters
# - Calculate area = length * width
# - Calculate perimeter = 2 * (length + width)
# - Pack both values into a tuple like (area, perimeter)
# - Return the tuple

# PROGRAM LOGIC
# - Ask user for length
# - Ask user for width
# - Call the function with user inputs
# - Unpack returned tuple into my_area and my_perimeter
# - Print the results clearly
# NAME: rectangle.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
def calculate_rectangle(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return (area, perimeter)

# PROGRAM LOGIC
# Test the function with fixed values
result = calculate_rectangle(5, 10)
print(result)
```
